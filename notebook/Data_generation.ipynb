#### Generate Survey Data

```python
# For reproducibility
np.random.seed(42)

# The total number of students in SS3 class.
num_records = 300

# Sample options for dataset columns.
genders = ['Female', 'Male']
departments = ['Arts', 'Science', 'Business']
home_languages = ['Native', 'English']
english_fluency = ['Not Fluent', 'Fluent', 'Very Fluent']
income_levels = ['Low Income', 'Middle Income', 'High Income']
education_levels = ['No Formal Education', 'Primary Education', 'Secondary Education' ,'Tertiary Education']
disabilities = ['Yes', 'No']
internet_access = ['Yes', 'No']
private_lessons = ['Yes', 'No']
extracurricular_activity = ['Yes', 'No']
resources_access = ['No access', 'Limited access', 'Adequate access', 'Full access']
disciplinary_actions = ['Yes', 'No']
class_participation = ['Rarely', 'Frequently']

# Parameters for the Age distribution.
age_mean = 17.0
age_std_dev = 1.23

# Generate random age values from a normal distribution.
random_age_values = norm.rvs(loc=age_mean, scale=age_std_dev, size=num_records)

# Create a DataFrame for Age.
education_df = pd.DataFrame(random_age_values, columns=['Age'])

# Round the values to get integer ages.
education_df['Age'] = education_df['Age'].round().astype(int)

# Ensure that no age value is less than 15.
education_df['Age'] = education_df['Age'].apply(lambda x: max(x, 15))

# Define the Gender proportion.
gender_proportions = {'Female': 0.590139, 'Male': 0.409861}
disability_proportions = {'Yes': 0.1, 'No': 0.9}

# Generate random gender values based on the given proportions.
education_df['Gender'] = np.random.choice(genders, size=num_records, p=[gender_proportions['Female'], gender_proportions['Male']])

# Define the Department proportion.
department_proportions = {'Science': 0.45, 'Business': 0.22, "Arts":0.33}

# Generate random department values based on the given proportions.
education_df['Department'] = np.random.choice(departments, size=num_records, p=[department_proportions["Arts"], department_proportions["Science"], department_proportions["Business"]])

# Generate other categorical data.
education_df['Home Language'] = np.random.choice(home_languages, size=num_records)
education_df['English Fluency'] = np.random.choice(english_fluency, size=num_records)
education_df['Parents Income'] = np.random.choice(income_levels, size=num_records)
education_df['Parents Highest Education Level'] = np.random.choice(education_levels, size=num_records)
education_df['Attendance Rate'] = np.round(np.random.uniform(50, 100, size=num_records)).astype(int)
education_df['Disability'] = np.random.choice(disabilities, size=num_records, p=[disability_proportions['Yes'], disability_proportions['No']])
education_df['Internet Access'] = np.random.choice(internet_access, size=num_records)
education_df['Private Lessons'] = np.random.choice(private_lessons, size=num_records)
education_df['Extracurricular Activity'] = np.random.choice(extracurricular_activity, size=num_records)
education_df['Resources Access'] = np.random.choice(resources_access, size=num_records)
education_df['Disciplinary Actions'] = np.random.choice(disciplinary_actions, size=num_records)
education_df['Class Participation'] = np.random.choice(class_participation, size=num_records)

# Add a unique student ID for each record.
education_df.insert(0, 'student_id', range(1, num_records + 1))

education_df.head()
```

#### Define means and Standard Deviations for the general subjects and specific subjects for each departments
to generate the assessment scores we focused on 9 subjects because most external exams in Nigeria require registration of 9 subjects.

```python
# Define means and standard deviations for general subjects.
general_subject_params = {
    'Maths': {'mean': 55, 'std': 9},
    'English': {'mean': 60, 'std': 10},
    'Civic_education': {'mean': 75, 'std': 8},
    'Economics': {'mean': 70, 'std': 8.5},
}

# Define means and standard deviations for specialized Science subjects.
science_subject_params = {
    'Biology': {'mean': 66, 'std': 10},
    'Chemistry': {'mean': 59, 'std': 14},
    'Physics': {'mean': 61, 'std': 12},
    'Geography': {'mean': 65, 'std': 11.5},
    'Agric': {'mean': 60, 'std': 8},
}

# Define means and standard deviations for specialized Arts subjects.
arts_subject_params = {
    'Lit_in_Eng': {'mean': 59, 'std': 10},
    'Government': {'mean': 59, 'std': 9},
    'CRS': {'mean': 70, 'std': 6},
    'Yoruba': {'mean': 63, 'std': 12},
    'History': {'mean': 55, 'std': 7},
}

# Define means and standard deviations for specialized Business subjects.
business_subject_params = {
    'Government': {'mean': 59, 'std': 9},
    'CRS': {'mean': 70, 'std': 6},
    'Accounting': {'mean': 64, 'std': 10},
    'Commerce': {'mean': 66, 'std': 9},
    'Marketing': {'mean': 68, 'std': 7},
}
```

### Define the Function to generate Assessment scores data For each Department.

```python
# Function to generate scores for general and specialized subjects for any department
def generate_department_scores(department, general_subjects, department_subjects):
    """
    Parameters:
    - department: The department for which to generate scores
    - general_subjects: Dictionary of general subjects (mean, std)
    - department_subjects: Dictionary of specialized subjects (mean, std)

    Returns:
    - DataFrame with scores for the specified department
    """
    # Filter the main DataFrame for the specified department
    dept_df = education_df[education_df['Department'] == department].copy()

    # Generate general subject scores for all students in the department
    for subject, params in general_subjects.items():
        score = norm.rvs(loc=params['mean'], scale=params['std'], size=len(dept_df)).clip(0, 97).round()

        # Adjust scores based on additional factors
        for i in range(len(dept_df)):
            if dept_df.iloc[i]['Resources Access'] in ['Adequate access', 'Full access']:
                score[i] += 5
            elif dept_df.iloc[i]['Resources Access'] not in ['Adequate access', 'Full access']:
                score[i] -= 7
            if dept_df.iloc[i]['Class Participation'] == 'Frequently':
                score[i] += 5
            elif dept_df.iloc[i]['Class Participation'] != 'Frequently':
                score[i] -= 8
            if dept_df.iloc[i]['Extracurricular Activity'] == 'Yes':
                score[i] += 5
            elif dept_df.iloc[i]['Extracurricular Activity'] == 'No':
                score[i] -= 8
            if dept_df.iloc[i]['Private Lessons'] == 'Yes':
                score[i] += 5
            elif dept_df.iloc[i]['Private Lessons'] == 'No':
                score[i] -= 8

        dept_df[subject] = score.clip(0, 93).round()

    # Generate specialized subject scores for all students in the department
    for subject, params in department_subjects.items():
        score = norm.rvs(loc=params['mean'], scale=params['std'], size=len(dept_df)).clip(0, 97).round()

        # Adjust scores based on additional factors
        for i in range(len(dept_df)):
            if dept_df.iloc[i]['Resources Access'] in ['Adequate access', 'Full access']:
                score[i] += 5
            elif dept_df.iloc[i]['Resources Access'] not in ['Adequate access', 'Full access']:
                score[i] -= 7
            if dept_df.iloc[i]['Class Participation'] == 'Frequently':
                score[i] += 5
            elif dept_df.iloc[i]['Class Participation'] != 'Frequently':
                score[i] -= 8
            if dept_df.iloc[i]['Extracurricular Activity'] == 'Yes':
                score[i] += 5
            elif dept_df.iloc[i]['Extracurricular Activity'] == 'No':
                score[i] -= 8
            if dept_df.iloc[i]['Private Lessons'] == 'Yes':
                score[i] += 5
            elif dept_df.iloc[i]['Private Lessons'] == 'No':
                score[i] -= 8

        dept_df[subject] = score.clip(0, 95).round()

    return dept_df
```

### Define function to calculate the Total Grade Point and Average Grade point

```python
# Function to calculate total grade points and average score for a given DataFrame.
def calculate_grade_points_and_average_score(df, subject_columns):
    # Calculate total grade points.
    df['Total_Grade_Points'] = df[subject_columns].sum(axis=1)

    # Calculate average score.
    df['Average_Score'] = df['Total_Grade_Points'] / len(subject_columns)

    # Retain only the desired columns.
    final_df = df[['student_id'] + subject_columns + ['Total_Grade_Points', 'Average_Score']]
    return final_df
```

Generate Subject data

```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Sample teacher IDs (assuming teacher IDs are 1 to 27)
teacher_ids = list(range(1, 28))

# Define subjects for each department
science_subjects = [
    "Mathathematics", "English", "Civic Education",
    "Economics", "Biology", "Chemistry",
    "Physics", "Geography", "Agric"
]

arts_subjects = [
    "Mathathematics", "English", "Civic Education",
    "Economics", "Government", "CRS",
    "Lit-in-English", "Yoruba", "History"
]

business_subjects = [
    "Mathathematics", "English", "Civic Education",
    "Economics", "Government", "CRS",
    "Accounting", "Commerce", "Marketing"
]

# Create a list for all subjects with their respective departments
subjects = []
for subject in science_subjects:
    subjects.append((subject, "Science"))
for subject in arts_subjects:
    subjects.append((subject, "Arts"))
for subject in business_subjects:
    subjects.append((subject, "Business"))

# Shuffle teacher_ids to ensure randomness
random.shuffle(teacher_ids)

# Ensure each teacher gets a distinct set of subjects
assigned_teacher_ids = []
for i in range(len(subjects)):
    assigned_teacher_ids.append(teacher_ids[i % len(teacher_ids)])

# Generate random subject data
data = {
    'subject_id': list(range(1, len(subjects) + 1)),  # Subject IDs from 1 to number of subjects
    'name': [subject[0] for subject in subjects],  # Subject names
    'teacher_id': assigned_teacher_ids,  # Distinct teacher IDs
    'last_update': [datetime.today() - timedelta(days=random.randint(1, 365)) for _ in range(len(subjects))]  # Random last update dates
}

# Create DataFrame
df = pd.DataFrame(data)

# Convert 'last_update' to date format
df['last_update'] = df['last_update'].dt.date

# Save to CSV
df.to_csv('subject_data.csv', index=False)
```
